```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt


X = [random.random() for i in range(10000)]


def f(x,l=1/8):
  if x>=0:
    return(l*math.exp(-l*x))
  else:
    return(0)
def F(x,l=1/8):
  if x>=0:
    return(1-math.e**(-l*x))
  else:
    return(0)
def Finv(y,l=1/8):
  if y>=0:
    x = (-1/l)*math.log(1-y)
    return(x)
  else:
    return(0)

X0 = np.linspace(0,70,1000)
Y0 = [f(x) for x in X0]
Y = [Finv(y) for y in X]

fig1 = plt.figure(figsize=(10,7))
ax = fig1.subplots()
ax.hist(Y,bins=30,density=True,alpha=1,rwidth=0.95,label='Simulaciones')
ax.plot(X0,Y0,'r-', label='Densidad')
ax.legend(fontsize=12)
plt.title('Punto C. Función (1)',fontsize=15)
plt.ylabel('$p_{X}(x)$',fontsize=15)
plt.xlabel('$D_X$',fontsize=15)
plt.show()
```

```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt


def f(x):
  if 0<=x<=2 or 4<=x<=6:
    return(1/4)
  else:
    return(0)

def Finv(y):
  if 0<=y<=1/2:
    x = 4*y
    return(x)
  elif 1/2<=y<=1:
    x = 4*y+2
    return(x)


X = [random.random() for i in range(10000)]
Y = [Finv(y) for y in X]
X0 = np.linspace(-1,7,1000)
Y0 = [f(x) for x in X0]

fig1 = plt.figure(figsize=(10,7))
ax = fig1.subplots()
ax.hist(Y,bins=30,density=True,alpha=1,rwidth=0.85,label='Simulaciones')
ax.plot(X0,Y0,'r-',label='Densidad')
ax.legend(fontsize=10)
plt.title('Punto C. Función (7)',fontsize=15)
plt.ylabel('$p_{X}(x)$',fontsize=15)
plt.xlabel('$D_X$',fontsize=15)
plt.show()
```

```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  if 0<=x<=10:
    return((1/150)*x)
  elif 10<=x<=30:
    return((1/10)-(1/300)*x)
  else:
    return(0)

def Finv(y):
  if 0<=y<=1/3:
    x = 10*math.sqrt(3*y)
    return(x)
  elif 1/3<=y<=1:
    x = ((-1/10) + math.sqrt((1/10)**2 - 4*(-1/600)*(-1/2-y)))/(2*(-1/600))
    return(x)

X = [random.random() for i in range(10000)]
Y = [Finv(y) for y in X]
X0 = np.linspace(-1,31,10000)
Y0 = [f(x) for x in X0]

fig1 = plt.figure(figsize=(10,7))
ax = fig1.subplots()
ax.hist(Y,bins=30,density=True,alpha=1,rwidth=0.85,label='Simulaciones')
ax.plot(X0,Y0,'r-',label='Densidad')
ax.legend(fontsize=12)
plt.title('Punto C. Función (8)',fontsize=15)
plt.ylabel('$p_{X}(x)$',fontsize=15)
plt.xlabel('$D_X$',fontsize=15)
plt.show()
```

```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
import sympy
from sympy import *
from sympy import symbols

w = symbols('w')

sd = 0.5**0.5
media = 0.3
X = sd * np.random.randn(10000) + media

def g(x):
  y = math.exp(x)
  return(y)

def f(x,sd=sd,m=media):
  if 0<=x:
    y = (1/math.sqrt(2*math.pi) * (1/sd) * (1/x)) * math.exp(-((math.log(x)-m)/(math.sqrt(2)*sd))**2)
  else:
    y = 0
  return(y)

Y = [g(y) for y in X]

X0 = np.linspace(-1,20,10000)
Y0 = [f(x) for x in X0]

fig1 = plt.figure(figsize=(10,7))
ax = fig1.subplots()
ax.hist(Y,bins=30,density=True,alpha=1,rwidth=0.85,label='Simulaciones')
ax.plot(X0,Y0,'r-',label='Densidad')
ax.legend(fontsize=12)
plt.title('Punto D. 1a',fontsize=15)
plt.ylabel('$p_{X}(x)$',fontsize=15)
plt.xlabel('$D_X$',fontsize=15)
plt.show()
```

####Función $5)$

```python
a = 5
b = 10

def h(x):
  y = math.sqrt(x) * math.sin(x) * (1/(x+1)**3)
  return(y)
def inte(tamaño):
  X = (b-a)*np.random.random(size = tamaño) + a #Uniforme[a,b]
  E_h = 0
  for x in X:
    E_h += (1/tamaño)*h(x)
  return(E_h)

print('{:^12s}{:^16s}'.format('r','Valor de integral'))
print('{:^12s}{:^14s}'.format('---','------'))
for r in [10**k for k in range(1,6)]:
  print('{:^12.0f}{:^16.16f}'.format(r,(b-a)*inte(r)))
```

```python
def Simpson(f,a,b,n=10000):
  h = float((b-a)/n)
  T = [a+i*h for i in range(n+1)]
  Valor = (h/3)*(f(T[0]) + f(T[n]))
  for k in range(1,n):
    if k%2==1:
      Valor += (h/3)*4*f(T[k])
    elif k%2==0:
      Valor += (h/3)*2*f(T[k])
  return(Valor)

print(f'Valor de la integral = {Simpson(h,a,b)}')
```
